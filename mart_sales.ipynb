```
x=df[['Item_MRP']]
y=df['Item_Outlet_Sales']
x.head()
```

```
import numpy as np
from sklearn.linear_model import LinearRegression
```

```
linreg = LinearRegression()
linreg.fit(x,y)
```

```
print(linreg.intercept_)
print(linreg.coef_)
```

```
dt.head()
x_test=dt[['Item_MRP']]
```

```
dt.head()
```

```
y_pred=linreg.predict(x_test)
```

```
res=dt.loc[:,['Item_Identifier','Outlet_Identifier']]
res.head()
```

```
res['Item_Outlet_Sales']=y_pred
res.set_index('Item_Identifier',inplace=True)
```
